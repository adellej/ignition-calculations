```python
from finite_volume import Diffuse
import numpy as np
from scipy import integrate
from scipy import interpolate
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm, Normalize
import scipy.linalg
import color
import cmasher as cmr
from kepdump import load
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
# read in data to dict:

Temp_data_hot = {}
x_data_hot = {}
z_data_hot = {}
rho_data_hot = {}
ycol_data_hot = {}


sizes = [0.0001,0.001, 0.005, 0.01,0.1, 0.3, 1.0, 1.5, 2.0]
lengths = [0.004, 0.04, 0.4, 1.2, 2.4, 4.0, 5.2, 6.4, 8.0]
names = ['run3_HS_size_1cm','run3_HS_size_1m', 'run3_HS_size_5m', 'run3_HS_size_10m', 'run3_HS_size_100m', 'run3_HS_size_300m', 'run3_HS_size_1km', 'run3_HS_size_1.5km', 'run3_HS_size_2km']
#names = ['run3_HS_size_1cm_cart','run3_HS_size_1m_cart', 'run3_HS_size_5m_cart', 'run3_HS_size_10m_cart', 'run3_HS_size_100m_cart', 'run3_HS_size_300m_cart', 'run3_HS_size_1km_cart', 'run3_HS_size_1.5km_cart', 'run3_HS_size_2km_cart']
temps =[1e8, 5e7]
qb=1

for i in range(len(names)):
    D = Diffuse(Nx=400,
        Nz=80,
        Lx=lengths[i]*1e5,
        Lz=600,
        Tcrust=1e7,
        Thotspot=1e7,
        Ttop=1e6,
         bb_flag='bb',
         Fbase=1.0e21,
        hsize=sizes[i]*400/lengths[i],
         tol=1e-6,
         op=1.0,
         maxIter=300,
         run_name="17480_1e8",
         rhoflag="fit",
         coord="cylindrical",
         kappa_path='/Users/adelle/Documents/ignition_calculations/cylindrical_coords/finite_volume_method/kappa.npz',
         dump_path=f'/Users/adelle/Documents/ignition_calculations/kepler/run3/h#298')
    with open(f"/Users/adelle/Documents/ignition_calculations/cylindrical_coords/Alex/heat/heat_substrate/heat/data/data_solutiontemp_run{names[i]}_cart_qb{qb}_hs{temps[0]}.txt") as f:
            lines = (line for line in f if not line.startswith("#"))
            Temp_data_hot[f'{names[i]}'] = np.loadtxt(lines, delimiter=" ", skiprows=1)
    x_data_hot[f'{names[i]}'] = D.x
    z_data_hot[f'{names[i]}'] = D.z
    rhovals2 = -1 * D.rhoz 
    rho_data_hot[f'{names[i]}'] = rhovals2
    ycol_data_hot[f'{names[i]}'] = integrate.cumtrapz(rhovals2[::-1], D.z[::-1], initial=1e-3)[::-1]
```
